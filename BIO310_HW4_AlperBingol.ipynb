1. How many rows and columns does the coding potential.csv have?

  57523 rows × 39 columns



```
X = df.drop('label',1) #features
y = df['label'] #target

print(X.shape) # Viewing the shape of X
print(y.shape) # Viewing the shape of y

X.head()
```

```
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y)
```

```
print(Xtrain.shape) # Viewing the shape of X
print(ytrain.shape) # Viewing the shape of y

print(Xtest.shape) # Viewing the shape of X
print(ytest.shape) # Viewing the shape of y
```

2. When you split the data into segments, how many samples (rows) are in the training and test data?

  Rows in training data: 43142
  Rows in test data: 14381

```
from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier(n_estimators=100, random_state=42)
model_rf.fit(Xtrain, ytrain)
```

3. Report the confusion matrices and classification reports for the training and test data

```
from sklearn.metrics import confusion_matrix, classification_report

ypred_train = model_rf.predict(Xtrain)
print('Train Confusion Matrix: ')
print(confusion_matrix(ytrain, ypred_train))
print('')

print('Train Classification report:')
print(classification_report(ytrain, ypred_train))
print('')
```

```
ypred = model_rf.predict(Xtest)
print('Test Confusion Matrix: ')
print(confusion_matrix(ytest, ypred))
print('')
print('Test Classification report:')
print(classification_report(ytest, ypred))
```

## 5. What are precision, recall and F1-score?

      Precision: Percentage of the items that the system labeled as positive and that are in fact positive. 
      Recall :  Percentage of the items actually present in the input that were correctly identified by the system.
      F1-Score: Since we have two measures (Precision and Recall) it helps to have a measurement that represents both of them. We calculate an F-measure which uses Harmonic Mean in place of Arithmetic Mean as it punishes the extreme values more. F1-Score is just an arithmetic mean of precision and recall since beta is 1.
